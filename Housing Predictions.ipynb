This Jupyter Notebook will use the California housing dataset from scikit-learn to make predictions regarding each column in the dataset. First, we will need to import Numpy, Pandas, MatplotLib and scikit-learns datasets to get started.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
```

Next, we will print out the dataset to see what kind of data is stored.

```python
housing = fetch_california_housing(as_frame=True)
df = pd.DataFrame(housing.data, columns=housing.feature_names)
print(df)
```

Looking at the dataset, we don't need the latitude and longitude columns.

```python
df = df.drop(['Latitude', 'Longitude'], axis=1) 
print(df)
```

Let us plot our data to see the relationship between the rows. 

```python
df.plot()
plt.show()
```

Next, we will choose the target variable and our features. The MedInc column will be our target value.

```python
X = df.drop(columns=["MedInc"])
y = df["MedInc"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, color='blue', label='Predicted vs Actual Values')

plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Predictions')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.legend()

plt.show()
```

As you can see from the Scatter Plot above, we have successfully made predictions for each column in the California dataset.
